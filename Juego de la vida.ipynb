```python
import copy
import random
import itertools
import time
import os

clear=lambda: os.system('clear') 

class JuegoDeLaVida(object):

    def __init__(self, filas, columnas):

        self.filas=filas
        self.columnas=columnas
        fila_viva=lambda: [random.randint(0, 1) for n in range(self.columnas)]
        self.juego=[fila_viva() for n in range(self.filas)]
        self.vida=1
        self.muerte=1

    def __str__(self):

        tabla=''
        for fila in self.juego:
            for celula in fila:
                tabla+=' ■ ' if celula else ' o '
            tabla+='\n'
        return tabla

    def evaluar(self, fila, columna):

        distancia=list(set(itertools.permutations([-1, -1, 1, 1, 0], 2)))
        existeTabla=lambda x, y: (x in range(self.filas) and y in range(self.columnas))

        total=0
        for f, c in distancia:
            if existeTabla(f+fila, c+columna):
                total+=self.juego[f+fila][c+columna]
        return total

    def test(self):

        tablaAux=copy.deepcopy(self.juego)
        self.vida=0
        self.muerte=0

        for f in range(self.filas):
            for c in range(self.columnas):
                total=self.evaluar(f, c)

                if (total<2 or total>3) and tablaAux[f][c]:
                    tablaAux[f][c]=0
                    self.muerte+=1
                elif total==3 and not tablaAux[f][c]:
                    tablaAux[f][c]=1
                    self.vida+=1

        self.juego=copy.deepcopy(tablaAux)

filas, columnas=int(input("Filas>> ")), int(input("Columnas>> "))

juegox=JuegoDeLaVida(filas, columnas)

iteraciones=0
while juegox.vida>0 or juegox.muerte>0:
    try:
        juegox.test()
        print(juegox)
        time.sleep(0.5)
        iteraciones+=1
    except KeyboardInterrupt:
        break
print("Total: ", iteraciones)   
        
```
